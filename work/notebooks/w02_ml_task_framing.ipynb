## 1. My lane as an ML task (type)

Task Type: Binary Classification / Scoring and Ranking.

Framing: We are treating this as a recommendation ranking problem driven by a binary classification engine. For every unique query-content interaction, the model outputs a probability score between 0 and 1 representing the likelihood of a user clicking that content item. These scores are then sorted in descending order to rank the results.

## 2. Target or proxy

Target Variable: Click occurrence (clicked: 1 for clicked, 0 for not clicked).

Proxy Justification: While the ultimate business goal is long-term user satisfaction and retention, immediate clicks serve as an explicit, high-signal proxy for historical relevance. We will use the difference between impressions and actual clicks in historical time-windows to train our target outcomes.

## 3. Success metric

Primary Evaluation Metric: NDCG (Normalized Discounted Cumulative Gain) and Mean Reciprocal Rank (MRR).

Why this metric works: Standard classification metrics like accuracy or F1-score fail here because search log datasets are extremely imbalanced (the vast majority of displayed search results never get clicked). NDCG and MRR explicitly account for position-based ranking quality, punishing the system heavily if a highly relevant item is buried at the bottom of the page where users rarely look.

## 4. The unit of analysis, as a real dataframe

Unit of Analysis: A unique Query-Content Pair per day/session context.

Dataframe Schema Design:
- query (string): The exact search keyword entered by a user.
- content_id (string): The unique platform identifier for the displayed asset item.
- imp_prev30 (int): Total historical impression baseline for this pair in the last 30 days.
- clicks_prev30 (int): Total historical click count for this pair in the last 30 days.
- target_clicked (int/binary): 1 if this interaction instance resulted in an actual user click, 0 otherwise.

## 5. Why ML beats a fixed rule here

Why ML Wins: A fixed rule (such as "always sort strictly by the highest historical click count") creates a devastating feedback loop: popular items get all the impressions, and new or long-tail content items can never surface. An ML model handles high-dimensional interaction signals simultaneously, allowing us to balance historical popularity trends with immediate query match relevance across millions of unique search combinations that a static if-statement cannot scale to resolve.
